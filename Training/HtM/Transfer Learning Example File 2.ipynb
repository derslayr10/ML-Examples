```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import itertools
import random
import os

import tensorflow as tf
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
from tensorflow.keras.utils import *

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.manifold import TSNE

from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = pd.read_csv("HtM_MAF Data_Final_Edited.csv", usecols = ["Corrosion", "Bare Metal", "Corrosion Prevention Treatment", "Routine Maintenance" , "Unscheduled Maintenance" , "Mission-Related Maintenance" , "Failure"])

#create encoder to easily shape data
labelEncoder = LabelEncoder()

#encode labels
df1["Corrosion"] = labelEncoder.fit_transform(df1["Corrosion"])
df1["Bare Metal"] = labelEncoder.fit_transform(df1["Bare Metal"])
df1["Corrosion Prevention Treatment"] = labelEncoder.fit_transform(df1["Corrosion Prevention Treatment"])
df1["Routine Maintenance"] = labelEncoder.fit_transform(df1["Routine Maintenance"])
df1["Unscheduled Maintenance"] = labelEncoder.fit_transform(df1["Unscheduled Maintenance"])
df1["Mission-Related Maintenance"] = labelEncoder.fit_transform(df1["Mission-Related Maintenance"])
df1["Failure"] = labelEncoder.fit_transform(df1["Failure"])

#create train and test sets
Train1 = pd.DataFrame(data = df1, columns = ["Bare Metal", "Corrosion Prevention Treatment", "Routine Maintenance" , "Unscheduled Maintenance" , "Mission-Related Maintenance" , "Failure"], copy = True)
Test1 = pd.DataFrame(data = df1, columns = ["Corrosion"], copy = True)

#create train and test data
train_features1, test_features1, train_labels1, test_labels1 = train_test_split(Train1, Test1, test_size = 0.25, random_state = random.seed(1))
```

```python
print("Train Features shape: ", train_features1.shape)
print("Train Labels shape: ", train_labels1.shape)
print("Test Features shape: ", test_features1.shape)
print("Test Labels shape: ", test_labels1.shape)
```

```python
model = load_model('tl_example_2.h5')
layer_list = list()
for layer in model.layers[1:]:
    layer.trainable = False
    layer_list.append(layer)
print(len(layer_list))
```

```python
model1 = Sequential()
model1.add(Dense(32, input_dim = train_features1.shape[1], activation = "relu", name = "layer1"))
model1.add(Dense(32, activation = "relu", name = "layer2"))
model1.add(Dropout(0.3))
model1.add(Dense(32, activation = "relu", name = "layer3"))
model1.add(Dropout(0.2))
model1.add(Dense(32, activation = "relu", name = "layer4"))
model1.add(Dropout(0.1))
model1.add(Dense(32, activation = "relu", name = "layer5"))
model1.add(Dense(32, activation = "relu", name = "layer6"))
model1.add(Dropout(0.2))
model1.add(Dense(32, activation = "relu", name = "layer7"))
model1.add(Dropout(0.3))
model1.add(Dense(32, activation = "relu", name = "layer8"))
model1.add(Dropout(0.1))
model1.add(Dense(32, activation = "relu", name = "layer9"))
model1.add(Dropout(0.2))
model1.add(Dense(32, activation = "relu", name = "layer10"))
model1.add(Dropout(0.2))
model1.add(Dense(32, activation = "relu", name = "layer11"))
model1.add(Dropout(0.2))
model1.add(Dense(32, activation = "relu", name = "layer12"))
model1.add(Dropout(0.1))
model1.add(layer_list[0])
model1.add(Dropout(0.1))
model1.add(layer_list[1])
model1.add(Dropout(0.1))
model1.add(layer_list[2])
model1.add(Dropout(0.1))
model1.add(layer_list[3])
model1.add(layer_list[4])
model1.add(layer_list[5])
model1.add(layer_list[6])
model1.add(layer_list[7])
model1.add(layer_list[8])
model1.add(Dropout(0.2))
model1.add(layer_list[9])
model1.add(Dropout(0.2))
model1.add(layer_list[10])
model1.add(Dropout(0.2))
model1.add(layer_list[11])
model1.add(Dropout(0.1))
model1.add(layer_list[12])
model1.add(Dropout(0.1))
model1.add(layer_list[13])
model1.add(Dropout(0.1))
model1.add(layer_list[14])
model1.add(Dropout(0.1))
model1.add(layer_list[15])

model1.compile(loss = "binary_crossentropy", optimizer = "adadelta", metrics = ["accuracy"])

model1.summary()
```

```python
history = model1.fit(train_features1, train_labels1, batch_size = 30, epochs = 200)
```
